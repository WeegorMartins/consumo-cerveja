```python
from sklearn.model_selection import train_test_split
```

```python
y = dados['consumo']
```

```python
X = dados[['temp_max', 'chuva', 'fds']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2811)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
modelo = LinearRegression()
```

```python
modelo.fit(X_train, y_train)
```

```python
print('R = {}'.format(modelo.score(X_train, y_train).round(2)))
```

```python
y_previsto = modelo.predict(X_test)
```

```python
print('R = %s' % metrics.r2_score(y_test, y_previsto).round(2))
```

```python
entrada = X_test[0:1]
entrada
```

```python
modelo.predict(entrada)[0].round(2)
```

## Interpretao dos Coeficientes Estimados

<p style='font-size: 20px; line-height: 2; margin: 10px 50px; text-align: justify;'>
<b>Intercepto</b>  Excluindo o efeito das variveis explicativas ($X_2=X_3=X_4=0$) o efeito mdio no Consumo de Cerveja seria de <b>5951,98 litros</b>.
</p>

<p style='font-size: 20px; line-height: 2; margin: 10px 50px; text-align: justify;'>
<b>Temperatura Mxima (C)</b>  Mantendo-se os valores de $X_3$ (Chuva) e $X_4$ (Final de Semana) constantes, o acrscimo de 1C na Temperatura Mxima gera uma variao mdia no Consumo de Cerveja de <b>684,74 litros</b>.
</p>

<p style='font-size: 20px; line-height: 2; margin: 10px 50px; text-align: justify;'>
<b>Chuva (mm)</b>  Mantendo-se os valores de $X_2$ (Temperatura Mxima) e $X_4$ (Final de Semana) constantes, o acrscimo de 1mm de Chuva gera uma variao mdia no Consumo de Cerveja de <b>-60,78 litros</b>.
</p>

<p style='font-size: 20px; line-height: 2; margin: 10px 50px; text-align: justify;'>
<b>Final de Semana (Sim/No)</b>  Mantendo-se os valores de $X_2$ (Temperatura Mxima) e $X_3$ (Chuva) constantes, o fato de o dia ser classificado como Final de Semana gera uma variao mdia no Consumo de Cerveja de <b>5401,08 litros</b>.
</p>
